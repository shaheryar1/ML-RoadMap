# Exploratory Data Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris=pd.read_csv('IRIS.csv')
iris.head()
```

```python
print("First five rows")
print(iris.head())
print("*********")
print("columns",iris.columns)
print("*********")
print("shape:",iris.shape)
print("*********")
print("Size:",iris.size)
print("*********")
print("no of samples available for each type") 
print(iris["type"].value_counts())
print("*********")
print(iris.describe())
```

```python
iris['type'].value_counts()
type(iris['type'].value_counts())
```

```python
iris['type'].value_counts().plot(kind='bar')
```

## Probability distribution analysis using Histogram

```python

#Histogram tell us which feature is good which is bad
sns.FacetGrid(iris,hue="type",size=3).map(sns.distplot,"petal_length").add_legend()
sns.FacetGrid(iris,hue="type",size=3).map(sns.distplot,"petal_width").add_legend()
sns.FacetGrid(iris,hue="type",size=3).map(sns.distplot,"sepal_length").add_legend()
sns.FacetGrid(iris,hue="type",size=3).map(sns.distplot,"sepal_width").add_legend()
plt.show()
```

```python
iris_setosa=iris.loc[iris["type"]=="Iris-setosa"]
iris_virginica=iris.loc[iris["type"]=="Iris-virginica"]
iris_versicolor=iris.loc[iris["type"]=="Iris-versicolor"]
```

## Features Comparison with scatter plot

```python
iris_setosa['sepal_width'].to_numpy()
```

```python
def plot_features(feature1_name,feature2_name):
    plt.scatter(iris_setosa[feature1_name],iris_setosa[feature2_name])
    plt.scatter(iris_virginica[feature1_name],iris_virginica[feature2_name])
    plt.scatter(iris_versicolor[feature1_name],iris_versicolor[feature2_name])
    plt.legend(['iris_setosa','iris_virginica','iris_versicolor'])
```

```python
# Only practice
a=[1,2,3,4,5,6,7,8]
b=[1,4,9,16,25,36,49,64]

c=[1,2,3,4,5,6,7,8]
d=[1,3,8,6,15,26,19,54]
plt.scatter(a,b)
plt.scatter(c,d)
plt.title('Sample project')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend(['a and b','c and d'])
```

```python
feature1_name = 'sepal_length'
feature2_name = 'sepal_width'
plot_features(feature1_name,feature2_name)
```

```python
feature1_name = 'petal_length'
feature2_name = 'petal_width'
plot_features(feature1_name,feature2_name)
```

### Outliers Analysis using Box Plot

```python
sns.boxplot(x="type",y="petal_width",data=iris)
plt.show()
```

```python
fig,ax=plt.subplots(2,2,figsize=(10, 10))
sns.boxplot(x="type",y="petal_width",data=iris,ax=ax[0][0])
sns.boxplot(x="type",y="petal_length",data=iris,ax=ax[0][1])
sns.boxplot(x="type",y="sepal_length",data=iris,ax=ax[1][0])
sns.boxplot(x="type",y="sepal_length",data=iris,ax=ax[1][1])

plt.show()
```
